# seaborn


```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
mtcars = pd.read_csv("mtcars.csv")
```

```python
mtcars.head(5)
```

```python
#lineplot
```

```python
sns.lineplot(x="gear",y="mpg",data=mtcars)
plt.show()
```

```python
sns.lineplot(x="gear",y="hp",hue="carb",style="cyl",data=mtcars)
plt.show()
```

```python
#boxplot
```

```python
sns.boxplot(x="gear",y="mpg",data=mtcars)
plt.show()
```

```python
sns.boxplot(x="gear",y="mpg", hue ="carb",data=mtcars,)
plt.show()
```

```python
sns.boxplot(x="gear",y="mpg", palette="rocket",data=mtcars,)
```

```python
sns.boxplot(x="gear",y="mpg", color="orange",data=mtcars,)
```

```python
#scatterplot
```

```python
sns.scatterplot(x="gear",y="mpg" ,data=mtcars,)
plt.show()
```

```python
plt.scatter(mtcars.mpg,mtcars.wt)
plt.show()
```

```python
sns.scatterplot(x="disp",y="drat" ,hue='hp',style='hp',data=mtcars,)
```

```python
#pairplot
```

```python
sns.pairplot(mtcars.iloc[:,0:3]) #histrogram of each column and 
#scater plot of each variable with respect toother coumns
```

```python
sns.pairplot(mtcars)
plt.show()




```

```python
#hist and distplot
```

```python
mtcars.head(5)
```

```python
sns.distplot(mtcars["mpg"])
plt.show()
```

```python
sns.distplot(mtcars["mpg"],hist=False,color="red")
plt.show()
```

```python
sns.distplot(mtcars["mpg"],color="green",kde=False,bins=20)
```

```python
sns.distplot(mtcars["mpg"],color="black",vertical=True)
```

```python
sns.jointplot(x="mpg",y="hp",color="red",data=mtcars)
```

```python
sns.jointplot(x="mpg",y="hp",color="red",kind="reg",data=mtcars)
```
