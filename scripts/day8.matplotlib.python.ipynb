```python
student ={"Bob":87,"Matt":56,"Sam":27}
```

```python
names = list(student.keys())
values =list(student.values())
plt.bar(names,values)
plt.title("Distubution of students marks")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```

```python
names = list(student.keys())
values =list(student.values())
plt.barh(names,values,color='g')
plt.title("Distubution of students marks")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```
